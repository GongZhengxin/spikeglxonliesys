```python
import numpy as np
import time

# 模拟300个stimulus的PSTH数据
test_data = {i: np.random.randn(350, 384).astype(np.float32) for i in range(300)}

# 测试deepcopy速度
start = time.perf_counter()
copy_data = dict(test_data)
elapsed = time.perf_counter() - start
print(f"Dict copy of 161MB: {elapsed*1000:.1f}ms")  # 通常50-150ms

# 如果dict中是对象引用，拷贝会更慢
start = time.perf_counter()
for _ in range(10):
    copy_data = {k: v.copy() for k, v in test_data.items()}
elapsed = (time.perf_counter() - start) / 10
print(f"Deep copy with numpy.copy(): {elapsed*1000:.1f}ms")  # 可能200-500ms
```
